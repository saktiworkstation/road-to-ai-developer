```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

df = pd.read_csv('/content/StudentPerformance-2.csv')

# Drop 'id_mhs' and 'nama' columns
df = df.drop(['id_mhs', 'nama'], axis=1)

# Separate features and target
X = df.drop('lulus_tepatwaktu', axis=1)
y = df['lulus_tepatwaktu']

# Encode target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Encode categorical features using OrdinalEncoder
categorical_cols = X.select_dtypes(include=['object']).columns
ordinal_encoder = OrdinalEncoder()
X_encoded = X.copy()
X_encoded[categorical_cols] = ordinal_encoder.fit_transform(X[categorical_cols])
```

```python
# Feature Selection using Filter Method (SelectKBest with ANOVA F-test)
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=10)
X_new = selector.fit_transform(X_encoded, y_encoded)

# Get selected feature names
mask = selector.get_support()
selected_features_filter = X_encoded.columns[mask]
print("Selected features using Filter method:")
print(selected_features_filter)

# Visualize selected features
for feature in selected_features_filter:
    plt.figure()
    if df[feature].dtype == 'object':
        sns.countplot(x=df[feature], hue=y)
        plt.title(f'Filter Method - {feature}')
        plt.xticks(rotation=45)
    else:
        sns.boxplot(x=y, y=df[feature])
        plt.title(f'Filter Method - {feature}')
    plt.tight_layout()
    plt.show()
```

```python
# Feature Selection using Wrapper Method (RFE with Logistic Regression)
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
rfe = RFE(estimator=model, n_features_to_select=10)
rfe.fit(X_encoded, y_encoded)

# Get selected feature names
mask = rfe.support_
selected_features_wrapper = X_encoded.columns[mask]
print("Selected features using Wrapper method:")
print(selected_features_wrapper)

# Visualize selected features
for feature in selected_features_wrapper:
    plt.figure()
    if df[feature].dtype == 'object':
        sns.countplot(x=df[feature], hue=y)
        plt.title(f'Wrapper Method - {feature}')
        plt.xticks(rotation=45)
    else:
        sns.boxplot(x=y, y=df[feature])
        plt.title(f'Wrapper Method - {feature}')
    plt.tight_layout()
    plt.show()
```

```python
# Feature Selection using Embedding Method (Feature Importance from Random Forest)
from sklearn.ensemble import RandomForestClassifier
import numpy as np

model = RandomForestClassifier()
model.fit(X_encoded, y_encoded)

# Get feature importances
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

# Select top 10 features
selected_indices = indices[:10]
selected_features_embedding = X_encoded.columns[selected_indices]
print("Selected features using Embedding method:")
print(selected_features_embedding)

# Visualize selected features
for feature in selected_features_embedding:
    plt.figure()
    if df[feature].dtype == 'object':
        sns.countplot(x=df[feature], hue=y)
        plt.title(f'Embedding Method - {feature}')
        plt.xticks(rotation=45)
    else:
        sns.boxplot(x=y, y=df[feature])
        plt.title(f'Embedding Method - {feature}')
    plt.tight_layout()
    plt.show()
```
