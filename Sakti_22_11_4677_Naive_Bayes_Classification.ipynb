```python
# Importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import *
import matplotlib.pyplot as plt
import seaborn as sns
```

# Binary Classification

```python
# Importing the dataset
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#X adalah variabel independen - berisi Age dan Estimated Salary
X = df.iloc[:, [2, 3]].values
#y adalah variabel dependen - berisi fitur Purchased
y = df.iloc[:, -1].values
```

```python
#menampilkan nilai x dan y
print('X : \n',X[:5])
print('\ny : \n',y[:5])
```

```python
# grafik dari nilai x dan y sebagai represntasi data
import matplotlib.pyplot as plt
# Create the plot
plt.scatter(X[:, 0], X[:, 1], c=y) # Use y for color coding
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.title("Scatter Plot of Age vs. Estimated Salary")
plt.colorbar(label="Purchased (0 or 1)") # Add colorbar label
plt.show()
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size = 0.20,
random_state = 0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Training the Naive Bayes model on the Training set
classifier_model_binary = GaussianNB()
classifier_model_binary.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier_model_binary.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
classifier_model_binary.predict_proba(X_test)
```

```python
def predict_one(age, salary):
  print('Prediction : ',classifier_model_binary.predict([[age, salary]]))
  print('Probability : ',classifier_model_binary.predict_proba([[age, salary]]))
```

```python
predict_one(30,87000)
```

```python
print(classification_report(y_test,y_pred,zero_division=0))
```

```python
# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
display_labels=classifier_model_binary.classes_)
disp.plot()
plt.show()
```

# Multiclass Classification

```python
#memasukkan data
df_train = pd.read_excel("/content/data_train.xlsx")
df_test = pd.read_csv("/content/data_test.csv")
```

```python
#menampilkan distribusi data pada data_train.xlsx
# Display data distribution for numerical features in df_train
df_train.describe()
# Display data distribution for categorical features in df_train
for col in df_train.select_dtypes(include=['object']).columns:
  print(f"\nDistribution of {col}:")
  print(df_train[col].value_counts())
# Visualize distributions using histograms and boxplots
df_train.hist(figsize=(10, 10))
plt.show()
df_train.plot(kind='box', subplots=True, layout=(2,3), figsize=(10, 10))
plt.show()
```

```python
X_train = df_train.drop(["Jurusan"], axis=1).values
y_train = df_train['Jurusan'].values
X_test = df_test.drop(["Jurusan"], axis=1).values
y_test = df_test['Jurusan'].values
```

```python
X_train
```

```python
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
classifier_model_multiclass = GaussianNB()
# Memasukkan data training pada fungsi klasifikasi naive bayes
classifier_model_multiclass.fit(X_train, y_train)
```

```python
#Melihat berapa kelas atau varibel target
import pandas as pd
# Load the training dataset
df_train = pd.read_excel("/content/data_train.xlsx")
# Get the unique values from the 'Jurusan' column
unique_classes = df_train['Jurusan'].unique()
# Print the unique classes (target variables)
print(f"Unique Classes (Target Variables):\n{unique_classes}")
print(f"\nNumber of unique classes: {len(unique_classes)}")
```

```python
y_predict = classifier_model_multiclass.predict(X_test)
print("Prediksi Naive Bayes : ",y_predict)
```

```python
#menentukan probabilitas hasil prediksi
classifier_model_multiclass.predict_proba(X_test)
```

```python
def predict_one(mat,ingg,indo,presis,presek):
  print('Predicted Jurusan : ',classifier_model_multiclass.predict([[mat,ingg,indo,presis,presek]]))
  print('Probability : ',classifier_model_multiclass.predict_proba([[mat,ingg,indo,presis,presek]]))
```

```python
predict_one(92,82,85,1,1)
```

```python
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
print(classification_report(y_test,y_predict,zero_division=0))
```

```python
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
display_labels=classifier_model_multiclass.classes_)
disp.plot()
plt.show()
```
