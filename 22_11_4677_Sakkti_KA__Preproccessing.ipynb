# Penugasan

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='Age', y='Fare')
plt.title("Scatterplot Sebelum membuang Outliers")

plt.subplot(1, 2, 2)
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot Sebelum membuang Outliers")

plt.tight_layout()
plt.show()
```

```python
# Membuang outliers dengan IQR
Q1 = df[['Age', 'Fare']].quantile(0.25)
Q3 = df[['Age', 'Fare']].quantile(0.75)
IQR = Q3 - Q1
df_filtered = df[~((df[['Age', 'Fare']] < (Q1 - 1.5 * IQR)) | (df[['Age', 'Fare']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df_filtered, x='Age', y='Fare')
plt.title("Scatterplot Setelah membuang Outliers")

plt.subplot(1, 2, 2)
sns.boxplot(data=df_filtered[['Age', 'Fare']])
plt.title("Boxplot Setelah membuang Outliers")

plt.tight_layout()
plt.show()
```

**Kesimpulan**

Nilai-nilai outlier dihilangkan dari himpunan data, sehingga menghasilkan distribusi yang lebih terkonsentrasi pada diagram sebar dan pengurangan dalam rentang diagram kotak.
